```python
import os
import glob
import pandas as pd
import datetime as dt
import re
```

```python
extension = 'csv'

all_filenames = [i for i in glob.glob('*.{}'.format(extension))]
df = pd.read_csv(all_filenames[0])
res = re.split('-', all_filenames[0])
date = res[2][0:8] 
datetime = []
for i in df.columns:
    next_datetime = date + " " + i
    datetime.append(next_datetime)
df = df.T
df = df.rename(columns = df.iloc[0])   
df = df.iloc[1:289]
for i in df.columns:
    next_datetime = date + " " + i
    datetime.append(next_datetime)
datetime = datetime[1:289]
df['Datetime'] = datetime
df.index = df['Datetime']
df.drop(columns = ['Datetime'], inplace = True)

```

```python
j = 0
for f in all_filenames:
    print(f)
    datetime = []
    res = re.split('-', f)
    date = res[2][0:8]    
    df_i = pd.read_csv(f)
    for i in df_i.columns:
        next_datetime = date + " " + i
        if 'Unnamed' in next_datetime:
            continue
        else:
            datetime.append(next_datetime)  
    datetime = datetime[1:289]
    df_i = df_i.T
    df_i = df_i.rename(columns = df_i.iloc[0])
    df_i = df_i.iloc[1:289]
    if len(df_i) != len(datetime):
        df_i = df_i.iloc[1:len(datetime)+1]
    #print('df', len(df_i))
    #print('dt', len(datetime))
    df_i['Datetime'] = datetime
    df_i.index = df_i['Datetime']
    df_i.drop(columns = 'Datetime', inplace = True)
    df = df.append(df_i)
    
df.index = pd.to_datetime(df.index)
df = df.sort_index()
```

```python
df.to_csv('supply-all-05.2018-05.20.csv')
```
